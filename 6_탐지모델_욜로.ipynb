```python
import time, cv2, numpy as np, pandas as pd 
from ultralytics import YOLO

# 환경 설정 및 공통 변수
IMG_PATH = "rsp.jpg"  # 분석 대상 이미지 파일 경로
MODELS = ["yolov8n.pt", "C:/githome/DL_rep/runs/detect/train4/weights/best.pt"]  # 비교할 YOLO 모델 파일들의 경로
CONF_THR = 0.25  # 모델 추론 시 탐지 결과를 확정하는 최소 신뢰도(Confidence) 임계값. 값이 낮을수록 더 많은 객체를 탐지할 수 있으나, 오탐지 확률 증가
IOU_THR = 0.70  # NMS(Non-Maximum Suppression) 시 사용되는 IOU(Intersection Over Union) 임계값. 겹치는 바운딩 박스를 제거하는 기준으로, 값이 낮을수록 겹치는 박스를 더 적극적으로 제거함
IMG_SIZE = 640  # 모델에 입력되는 이미지의 크기(가로, 세로). 모든 모델에 동일하게 적용하여 공정한 비교를 위함

def run_once(model_path, img_path):
    # YOLO 모델 로드. 모델 경로를 인자로 받아 모델 객체를 생성
    model = YOLO(model_path)

    # 추론 (모델 실행)
    t0 = time.perf_counter()  # 추론 시작 시간 기록
    # model() 함수로 추론 실행. verbose=False로 설정하여 추론 과정의 상세 출력은 생략
    res = model(img_path, imgsz=IMG_SIZE, conf=CONF_THR, iou=IOU_THR, verbose=False)[0]
    latency_ms = (time.perf_counter() - t0) * 1000  # 추론에 걸린 시간(밀리초) 계산

    # 결과 시각화 및 저장
    vis = res.plot()  # 추론 결과를 시각화(바운딩 박스, 라벨 등)하여 NumPy 배열(BGR 형식)로 반환
    out_name = model_path.replace(".pt", "_result.png")  # 결과 이미지 저장 파일명 생성
    cv2.imwrite(out_name, vis)  # 시각화된 이미지를 파일로 저장

    # 기본 통계 정보 추출
    num = int(len(res.boxes))  # 탐지된 객체(바운딩 박스)의 총 개수
    # 탐지된 객체들의 신뢰도 평균 계산. 객체가 없으면 0.0 반환
    mean_conf = float(res.boxes.conf.mean().item()) if num > 0 else 0.0
    # 탐지된 객체의 클래스 ID 목록 추출
    classes = [int(c) for c in res.boxes.cls.tolist()] if num > 0 else []
    # 클래스 ID를 사람이 읽을 수 있는 이름으로 변환
    class_names = [res.names[c] for c in classes]
    # 각 클래스별 탐지된 객체 수 계산
    uniq, counts = np.unique(class_names, return_counts=True)
    class_breakdown = dict(zip(uniq.tolist(), counts.tolist()))

    # 마스크(세그멘테이션) 면적 비율 계산 (모델이 세그멘테이션을 지원할 경우에만)
    mask_ratio = None
    if getattr(res, "masks", None) is not None and res.masks is not None:
        m = res.masks.data  # 세그멘테이션 마스크 데이터(Tensor)
        area = float(m.sum().item())  # 마스크 영역의 총 픽셀 수 계산
        H, W = m.shape[-2], m.shape[-1]  # 마스크의 높이와 너비
        mask_ratio = area / (H * W)  # 마스크 영역이 전체 이미지에서 차지하는 비율 계산

    # 결과를 딕셔너리 형태로 반환
    return {
        "model": model_path,
        "latency_ms": round(latency_ms, 2),  # 추론 시간
        "detections": num,  # 탐지된 객체 수
        "mean_conf": round(mean_conf, 4),  # 평균 신뢰도
        "classes": class_breakdown,  # 클래스별 탐지 객체 수
        "mask_ratio": round(mask_ratio, 6) if mask_ratio is not None else None,  # 마스크 비율
        "out_image": out_name
    }

# 메인 실행부
rows = [run_once(m, IMG_PATH) for m in MODELS]  # MODELS 리스트의 각 모델에 대해 run_once 함수를 실행
df = pd.DataFrame(rows, columns=[
    "model", "latency_ms", "detections", "mean_conf", "classes", "mask_ratio", "out_image"
])  # 반환된 딕셔너리 리스트를 Pandas DataFrame으로 변환

print(df.to_string(index=False))  # 데이터프레임을 인덱스 없이 깔끔한 문자열 표 형태로 출력
```
